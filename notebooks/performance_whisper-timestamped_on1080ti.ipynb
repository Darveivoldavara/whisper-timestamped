# Производительность **whisper-timestamped** на текущем сервере с **1080Ti**

### Статистика по **2 584** записям.

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook_connected"
```

```python
durations_comparsion = pd.read_csv(
    '/home/vladislav/ds/projects/a-iceberg/whisper-timestamped/output/performance_results.csv'
)

trace1 = go.Scatter(
    x=durations_comparsion.index,
    y=durations_comparsion['duration_audio'],
    mode='lines',
    name='Длительность аудио',
    line=dict(width=2)
)
trace2 = go.Scatter(
    x=durations_comparsion.index,
    y=durations_comparsion['duration_transcript'],
    mode='lines',
    name='Длительность распознавания',
    line=dict(width=2)
)

median_duration_audio = durations_comparsion['duration_audio'].median()
median_duration_transcript = durations_comparsion['duration_transcript'].median()

trace3 = go.Scatter(
    x=durations_comparsion.index,
    y=[median_duration_audio] * len(durations_comparsion),
    mode='lines',
    name='Медианная длительность аудио',
    line=dict(width=4)
)
trace4 = go.Scatter(
    x=durations_comparsion.index,
    y=[median_duration_transcript] * len(durations_comparsion),
    mode='lines',
    name='Медианная длительность распознавания',
    line=dict(width=4)
)

fig = go.Figure(data=[trace1, trace2, trace3, trace4])
fig.update_layout(
    width=1600,
    height=700,
    xaxis_title='Записи',
    yaxis_title='Секунды',
    legend_orientation='h',
    legend=dict(
        font=dict(
            size=15
        )
    ),
    title=dict(
        text='Длительности аудиозаписей и их распознавания',
        x=.5
    )
)
fig.show()
```

```python
trace1 = go.Scatter(
    x=durations_comparsion.index,
    y=durations_comparsion['performance_ratio'],
    mode='lines',
    name='Производительность распознавания',
    line=dict(width=2)
)

median_performance_ratio = durations_comparsion['performance_ratio'].median()

trace2 = go.Scatter(
    x=durations_comparsion.index,
    y=[median_performance_ratio] * len(durations_comparsion),
    mode='lines',
    name='Медианная производительность распознавания',
    line=dict(width=4)
)

fig = go.Figure(data=[trace1, trace2])
fig.update_layout(
    width=1600,
    height=700,
    xaxis_title='Записи',
    yaxis_title='Производительность, запись/распознавание',
    legend_orientation='h',
    legend=dict(
        font=dict(
            size=15
        )
    ),
    title=dict(
        text='Производительность распознавания whisper-timestamped',
        x=.5
    )
)
fig.show()
```

### В итоге медианная производительность составляет **0.06**, то есть в среднем запись распознается в **16** раз быстрее, чем длится сама.
### Колеблется производительность довольно сильно, от **0.02** (в **50** раз быстрее) до **0.45** (в **2** раза быстрее).

### По результатам предыдущего исследования текущая медианная длительность файлов для расшифровки в сутки — **20 376** мин (**340** ч) — данные со 02.01.2024 по 23.01.2024.
### То есть в среднем записи за сутки на текущих GPU-мощностях распознаются за **1 223** мин (**20** ч).

### В худшем случае (**533** часа записей в сутки при производительности **0.45**) аудио за день будут расшифрованы за **238** часов.
### В лучшем (**233** часа записей в сутки при производительности **0.02**) — за **5** часов.
